```python
# import files and upload them as dataframe
Banks_2019 = pd.read_csv("Resources/ALL_2019.csv", encoding = "ISO-8859-1")
Loans_Leases_2019_1 = pd.read_csv("Resources/All_Reports_20190331_Net Loans and Leases.csv")
Loans_Leases_2019_2 = pd.read_csv("Resources/All_Reports_20190630_Net Loans and Leases.csv")
```

```python
# Rename columns
Loans_Leases_2019_1.rename(columns={'cert':'ID', 'fed_rssd':'RSSDID'}, inplace=True)
Loans_Leases_2019_2.rename(columns={'cert':'ID', 'fed_rssd':'RSSDID'}, inplace=True)
```

```python
#Keep selected columns
Loans_Leases_2019_q1 = Loans_Leases_2019_1[['ID', 'RSSDID', 'lnlsnet', 'lnre', 'lnrecons', 'lnrenres', 'LNAUTO', 'lnreres', 'lncon']]
Loans_Leases_2019_q2 = Loans_Leases_2019_2[['ID','RSSDID','lnlsnet', 'lnre', 'lnrecons', 'lnrenres','LNAUTO', 'lnreres', 'lncon']]
```

```python
# Rename columns
Loans_Leases_2019_q1 = Loans_Leases_2019_q1.rename(columns={"lnlsnet": "TOTAL_LOANS_1","lnre": "REAL_ESTATE_LOANS_1",
                                                  "lnrecons": "CONSTRUCTION_LOANS_1", "lnrenres": "NONRESIDENTIAL_LOANS_1", 
                                                           "LNAUTO": "AUTO_LOANS_1", "lnreres": "RESIDENTIAL_LOANS_1", "lncon": "INDIVIDUAL_LOANS_1"})
Loans_Leases_2019_q2 = Loans_Leases_2019_q2.rename(columns={"lnlsnet": "TOTAL_LOANS_2", "lnre": "REAL_ESTATE_LOANS_2",
                                                  "lnrecons": "CONSTRUCTION_LOANS_2", "lnrenres": "NONRESIDENTIAL_LOANS_2",
                                                           "LNAUTO": "AUTO_LOANS_2", "lnreres": "RESIDENTIAL_LOANS_2", "lncon": "INDIVIDUAL_LOANS_2"})
```

```python
# Merge four data frames
Loans_Leases = pd.merge(Loans_Leases_2019_q1,Loans_Leases_2019_q2, on=("ID","RSSDID"), how="right")
#Loans_Leases = pd.merge(Loans_Leases_1,Loans_Leases_2, on="ID", how="right")
# Rename columns
Loans_Leases.rename(columns={'RSSDID_x': 'RSSDID'}, inplace=True)
Loans_Leases.head()
```

```python
Loans_Leases['TOTAL_LOANS'] = Loans_Leases['TOTAL_LOANS_1'] + Loans_Leases['TOTAL_LOANS_2']
Loans_Leases['REAL_ESTATE_LOANS'] = Loans_Leases['REAL_ESTATE_LOANS_2'] + Loans_Leases['REAL_ESTATE_LOANS_2']
Loans_Leases['CONSTRUCTION_LOANS'] = Loans_Leases['CONSTRUCTION_LOANS_1'] + Loans_Leases['CONSTRUCTION_LOANS_2']
Loans_Leases['NONRESIDENTIAL_LOANS'] = Loans_Leases['NONRESIDENTIAL_LOANS_1'] + Loans_Leases['NONRESIDENTIAL_LOANS_2']
Loans_Leases['AUTO_LOANS'] = Loans_Leases['AUTO_LOANS_1'] + Loans_Leases['AUTO_LOANS_2']
Loans_Leases['RESIDENTIAL_LOANS'] = Loans_Leases['RESIDENTIAL_LOANS_1'] + Loans_Leases['RESIDENTIAL_LOANS_2']
Loans_Leases['INDIVIDUAL_LOANS'] = Loans_Leases['INDIVIDUAL_LOANS_1'] + Loans_Leases['INDIVIDUAL_LOANS_2']
```

```python
Net_Loans= Loans_Leases
Net_Loans.head()
```

```python
Loans_2019 = Net_Loans[['ID','RSSDID','TOTAL_LOANS','REAL_ESTATE_LOANS','CONSTRUCTION_LOANS','NONRESIDENTIAL_LOANS','AUTO_LOANS','RESIDENTIAL_LOANS','INDIVIDUAL_LOANS']]

Loans_2019.head()
```
